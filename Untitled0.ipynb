```python
# Instalar dependência para leitura de arquivos docx
!pip install python-docx

# Monta o Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importações necessárias
import os
from docx import Document
import pandas as pd

# Caminho para a pasta com as entrevistas
pasta = "/content/drive/MyDrive/Doutorado/Analises/Entrevistas_Transcritas"

# Função para ler o conteúdo de um arquivo .docx
def ler_docx(caminho):
    doc = Document(caminho)
    return "\n".join([p.text for p in doc.paragraphs if p.text.strip() != ""])

# Lê todos os arquivos e guarda num DataFrame
dados = []
for arquivo in os.listdir(pasta):
    if arquivo.endswith(".docx"):
        caminho_arquivo = os.path.join(pasta, arquivo)
        texto = ler_docx(caminho_arquivo)
        dados.append({"arquivo": arquivo, "texto": texto})

df_entrevistas = pd.DataFrame(dados)

# Exibe uma amostra
df_entrevistas.head()
```

```python
# Caminho para o CSV das correspondências (se já tiver enviado para o Drive, atualize o caminho)
csv_path = "/content/drive/MyDrive/Doutorado/Analises/correspondencias_especies.csv"

# Lê o CSV com os nomes-tipo
df_corr = pd.read_csv(csv_path)
lista_tipos = df_corr["Tipo"].dropna().unique()

# Função para buscar nomes-tipo no texto da entrevista
import re

def identificar_especies(texto):
    encontrados = []
    texto_lower = texto.lower()
    for tipo in lista_tipos:
        tipo_lower = tipo.lower()
        if re.search(rf'\b{re.escape(tipo_lower)}\b', texto_lower):
            encontrados.append(tipo)
    return list(set(encontrados))  # remove duplicados

# Aplica a função ao DataFrame das entrevistas
df_entrevistas["especies_mencionadas"] = df_entrevistas["texto"].apply(identificar_especies)

# Exibe amostra
df_entrevistas[["arquivo", "especies_mencionadas"]].head()
```

```python
import re
import pandas as pd

# Lista base de locais relevantes nos territórios aquáticos da pesquisa
lista_locais = [
    "Jubim", "Bragança", "Marudá", "Salinas", "Salgado", "Amazonas", "Amapá",
    "Caeté", "Marajó", "Ajuruteua", "México", "Guiana", "Capanema", "Vila", "Praia",
    "Baía", "Manguezal", "Estuário", "Foz", "Rio", "Mato", "Maresia", "Mar"
]

# Função para identificar locais mencionados em um trecho de texto
def extrair_locais(texto):
    locais_identificados = []
    texto_lower = texto.lower()
    for local in lista_locais:
        if re.search(rf'\b{re.escape(local.lower())}\b', texto_lower):
            locais_identificados.append(local)
    return list(set(locais_identificados))  # remove repetições

# Varredura das entrevistas, frase por frase
registros_locais = []

for _, linha in df_entrevistas.iterrows():
    frases = re.split(r'[.!?]', linha["texto"])
    for frase in frases:
        locais = extrair_locais(frase)
        if locais:
            registros_locais.append({
                "arquivo": linha["arquivo"],
                "trecho": frase.strip(),
                "locais_mencionados": locais
            })

# Constrói o DataFrame final com os trechos que mencionam lugares
df_locais = pd.DataFrame(registros_locais)

# Visualiza os primeiros 10 registros
df_locais.head(10)
```

```python
# Núcleos espaciais relevantes no vocabulário local
nucleos_espaciais = [
    'rio', 'mangue', 'mar', 'ilha', 'praia', 'canal', 'costa', 'foz', 'margem',
    'saída', 'boca', 'beira', 'barranco', 'caminho', 'rede', 'vila', 'estaleiro',
    'poço', 'porto', 'fundão', 'salgado', 'baía', 'enseada', 'banco', 'cabeceira'
]

# Regex que extrai expressões locativas com núcleos válidos
regex_locativos_filtrados = re.compile(
    r'\b(?:na|no|em|pra|atrás|beira|boca|pé|saída|foz|margem|ilha|beira|banda|ponta|lado|fundo|costeira)\s(?:da|do|de)?\s?[a-zçãéêôí]+(?:\s(?:do|da|de)\s[a-zçãéêôí]+)?',
    re.IGNORECASE
)

# Aplicar apenas se o núcleo final da expressão estiver entre os aceitáveis
locativos_refinados = []

for _, linha in df_entrevistas.iterrows():
    matches = regex_locativos_filtrados.findall(linha["texto"])
    for m in matches:
        final_word = m.strip().split()[-1].lower()
        if final_word in nucleos_espaciais:
            locativos_refinados.append({
                "arquivo": linha["arquivo"],
                "expressao": m.strip()
            })

df_locativos_refinados = pd.DataFrame(locativos_refinados).drop_duplicates()
df_locativos_refinados.head(15)
```

```python
palavras_temporais = [
    # Meses e estações
    'janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
    'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro',
    'verão', 'inverno',

    # Ciclos com expressão
    'tempo de chuva', 'época da chuva', 'época de seca', 'tempo da seca',
    'época do salgado', 'época do peixe', 'época da sardinha',
    'época do bagre', 'época da maré', 'época da lua', 'época da cheia',

    # Outros relevantes
    'quando a água engrossa', 'quando a água baixa', 'quando enche',
    'quando começa a chover', 'quando a maré muda', 'na vazante', 'na cheia'
]


# Usa regex para encontrar expressões compostas ou exatas
def extrair_temporalidade_refinada(texto):
    texto_lower = texto.lower()
    encontrados = []
    for termo in palavras_temporais:
        if termo in texto_lower:
            encontrados.append(termo)
    return list(set(encontrados))

# Nova extração
registros_temporais_refinados = []

for _, linha in df_entrevistas.iterrows():
    frases = re.split(r'[.!?]', linha["texto"])
    for frase in frases:
        encontrados = extrair_temporalidade_refinada(frase)
        if encontrados:
            registros_temporais_refinados.append({
                "arquivo": linha["arquivo"],
                "trecho": frase.strip(),
                "temporalidades": ", ".join(encontrados)
            })

df_temporalidades = pd.DataFrame(registros_temporais_refinados).drop_duplicates()
df_temporalidades.head(15)
```

```python
# Lista de variações ambientais reconhecíveis
condicoes_ecologicas = [
    # Salinidade e textura da água
    "água fina", "água grossa", "água pesada", "água salgada", "água doce", 'misturando', 'adoçando', 'salgando', 'repiquete', 'subindo', 'descendo'
    "água verde", "água preta", "água barrenta", "água suja", "água clara",

    # Batimetria e profundidade
    "fundo", "raso", "fundão", "profundo", "na beira", "na costa", "na borda", 'baixo' 'cima'

    # Maré e ciclos hídricos
    "maré cheia", "maré seca", "vazante", "preamar", "enchente", "baixamar", 'pracuema', 'lanço', 'choca', 'morta', 'lançando', 'parando', 'baixola', 'preola'

    # Corrente e fluxo
    "corrente forte", "corrente fraca", "a água tá correndo", "água parada",

    # Tipo de fundo
    "lama", "areia", "cascalho", "pedra", "banco", "coroa", "muito fundo", "rasinho"
]

# Função de extração
def extrair_condicoes_ecologicas(texto):
    texto_lower = texto.lower()
    encontrados = [e for e in condicoes_ecologicas if e in texto_lower]
    return list(set(encontrados))

# Busca por frases que contenham essas expressões
registros_ecologicos = []

for _, linha in df_entrevistas.iterrows():
    frases = re.split(r'[.!?]', linha["texto"])
    for frase in frases:
        encontrados = extrair_condicoes_ecologicas(frase)
        if encontrados:
            registros_ecologicos.append({
                "arquivo": linha["arquivo"],
                "trecho": frase.strip(),
                "condicoes_ecologicas": ", ".join(encontrados)
            })

df_ambiente = pd.DataFrame(registros_ecologicos).drop_duplicates()
df_ambiente.head(30)
```

```python
import pandas as pd
import re

# Inicializa base com frases e espécies mencionadas
fluxos = []

for _, linha in df_entrevistas.iterrows():
    frases = [f.strip() for f in re.split(r'[.!?]', linha["texto"]) if len(f.strip()) > 30]
    especies = linha.get("especies_mencionadas", [])
    if isinstance(especies, list) and especies:
        for frase in frases:
            for especie in especies:
                fluxos.append({
                    "arquivo": linha["arquivo"],
                    "trecho": frase,
                    "especie": especie
                })

df_fluxo_base = pd.DataFrame(fluxos)

# Mescla com locais compostos
df_fluxo_1 = df_fluxo_base.merge(df_locativos_refinados, on=["arquivo", "trecho"], how="left")
df_fluxo_1.rename(columns={"expressao": "local_composto"}, inplace=True)

# Mescla com temporalidades
df_fluxo_2 = df_fluxo_1.merge(df_temporalidades, on=["arquivo", "trecho"], how="left")

# Mescla com condições ambientais
df_fluxo_final = df_fluxo_2.merge(df_ambiente, on=["arquivo", "trecho"], how="left")

# Normaliza valores booleanos (preenche False para NaN)
variaveis_ambientais = ['salinidade', 'cor_agua', 'batimetria', 'mare', 'corrente', 'substrato']
for col in variaveis_ambientais:
    if col in df_fluxo_final.columns:
        df_fluxo_final[col] = df_fluxo_final[col].fillna(False)

# Visualiza o resultado
df_fluxo_final.head(150)
```
